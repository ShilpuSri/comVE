```python
import itertools
import difflib
import pandas as pd
```

```python
def get_reasons(result):
    sent_orig=""
    result1=[]
    for i in range(len(result)):
        if i!=0 and i!=1 and '======================================== SAMPLE' not in result[i] and '<|endoftext|\n' not in result[i] and '<|endoftext|>' not in result[i]:
            if len(result[i].split()) > 4:
                result1.append(result[i].rstrip('\n'))
    sent_orig=str(result[1].rstrip('\n'))
    orig.append(sent_orig + ' <eol>')
    temp=set(result1)
    result1=list(temp)
    x=(difflib.get_close_matches(sent_orig, result1, n=3))
    rest=list(set(result1) - set(x))
    len1=len(x)
    if len1 == 3:
        reason1.append(x[0] + ' <eol>')
        reason2.append(x[1] + ' <eol>')
        reason3.append(x[2] + ' <eol>')
    elif len1 == 2:
        reason1.append(x[0] + ' <eol>')
        reason2.append(x[1] + ' <eol>')
        if len(rest) >= 1:
            reason3.append(rest[0] + ' <eol>')
        else:
            reason3.append(' <eol>')
    elif len1 == 1:
        reason1.append(x[0] + ' <eol>')
        if len(rest) >= 2:
            reason2.append(rest[0] + ' <eol>')
            reason3.append(rest[1] + ' <eol>')
        else:
            reason2.append(' <eol>')
            reason3.append(' <eol>')
    elif len1 == 0:
            if len(rest) >= 3:
                reason1.append(rest[0] + ' <eol>')
                reason2.append(rest[1] + ' <eol>')
                reason3.append(rest[2] + ' <eol>')
            elif len(rest) >= 2:
                reason1.append(rest[0] + ' <eol>')
                reason2.append(rest[1] + ' <eol>')
                reason3.append(' <eol>')
            elif len(rest) >= 1:
                reason1.append(rest[0] + ' <eol>')
                reason2.append(' <eol>')
                reason3.append(' <eol>')
            else:
                reason1.append(' <eol>')
                reason2.append(' <eol>')
                reason3.append(' <eol>')
```

```python
orig = []
reason1=[]
reason2=[]
reason3=[]
for i in range(1000):
    with open("output.txt") as fp:
        str1 = "Test Sample: %d" %(i+1)
        str2 = "Test Sample: %d" %i
        result = list(itertools.takewhile(lambda x: str1 not in x, itertools.dropwhile(lambda x: str2 not in x, fp)))
        get_reasons(result)
```

```python
df = pd.DataFrame(columns=['FalseSent', 'GPT2 Reason 1', 'GPT2 Reason 2', 'GPT2 Reason 3'])
```

```python
df['FalseSent'] = orig
df['GPT2 Reason 1'] = reason1
df['GPT2 Reason 2'] = reason2
df['GPT2 Reason 3'] = reason3
```
